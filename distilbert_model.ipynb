```python
import pandas as pd
import json
import sys
import os
import xmltodict
```

```python
fiant_data = []
for (r,d,f) in os.walk('xml'):
  if f:
    print(r,d,f)
    for fi in f:
      print(fi)
      if '.xml' not in fi:
        continue
      with open(r+'/'+fi, 'r') as x:
        i = x.read()
      t = xmltodict.parse(i)
      fiant_data.append((r+'/'+fi,t))
```

```python
fiant_data[0]
```

```python
import re
fiant_text = []
sources=[]
annotation_pattern = r'\<.*\>|,'
for f in fiant_data:
  sources.append(f[0])
  text = f[1]['TEI']['text']['body']['div']['div']['ab']
  if type(text) is list:
    text = [x for x in text if x is not None]
    text = "".join(text)
    text = re.sub(annotation_pattern, '', text)
  else:
    text = re.sub(annotation_pattern, '', text)
  fiant_text.append(text)
```

```python
pd.DataFrame({'src': sources, 'col': fiant_text}).to_csv('Fiant_data.csv', encoding='utf-8', index=None)
```

```python
import random
random.seed(1739)
data = pd.read_csv('Chancery Pleadings Metadata + Transcripts - Sheet1.csv',usecols=['LONG_TITLE','SHORT_TITLE','PLEADING_TEXT'])
data = list(data['PLEADING_TEXT'])
fiant_text = fiant_text + data
random.shuffle(fiant_text)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("dslim/distilbert-ner")

fiant_tokenised = []
for i, val in enumerate(fiant_text):
  # source = sources[i]
  fiant_tokenised.append(tokenizer.tokenize(val))
```

```python
from transformers import AutoModelForTokenClassification, pipeline

model = AutoModelForTokenClassification.from_pretrained("bert-base-uncased")
pipe = pipeline('ner', model=model, tokenizer=tokenizer)
results = pipe(fiant_tokenised[0])
```

```python
fiant_tokens = [cell for row in fiant_tokenised for cell in row]
pd.DataFrame({'col': fiant_tokens, 'labels': ['']*len(fiant_tokens)}).to_csv('Fiant_data_tokens.csv', encoding='utf-8', index=None)
```

```python
format_data = [{'tokens':[item['word'] for item in row], 'ner_tags':[item['entity'] for item in row]} for row in data]
labels = ['O'] + list(set([item for row in format_data for item in row['ner_tags']]))
id2label = {i:label for i, label in enumerate(labels)}
label2id = {label:i for i, label in enumerate(labels)}
format_data = [{'tokens':row['tokens'], 'ner_tags_str': row['ner_tags'], 'ner_tags':[label2id[item] for item in row['ner_tags']]} for row in format_data]
format_data = pd.DataFrame(format_data)
```

```python
from sklearn.model_selection import train_test_split
from datasets import Dataset, DatasetDict, Sequence, ClassLabel


# format_data = format_data.iloc[0:1400]
train_data, test_data = train_test_split(format_data, test_size=0.25)
test_data, eval_data = train_test_split(test_data, test_size=0.4)
train_data = Dataset.from_pandas(train_data)
train_data = train_data.cast_column('ner_tags', Sequence(ClassLabel(names=labels)))
test_data = Dataset.from_pandas(test_data)
test_data = test_data.cast_column('ner_tags', Sequence(ClassLabel(names=labels)))
eval_data = Dataset.from_pandas(eval_data)
eval_data = eval_data.cast_column('ner_tags', Sequence(ClassLabel(names=labels)))
data = DatasetDict({"train":train_data, "test":test_data, "eval":eval_data})
```

```python
# # Load model directly
# from transformers import AutoModel, AutoTokenizer, pipeline, AutoModelForTokenClassification
# tokenizer = AutoTokenizer.from_pretrained("emanjavacas/MacBERTh")
# model = AutoModelForTokenClassification.from_pretrained("emanjavacas/MacBERTh")


# pipe = pipeline('ner', model=model, tokenizer=tokenizer)
```

```python
from transformers import AutoTokenizer

model_checkpoint = "dslim/distilbert-ner"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
inputs = data['train'][0]['tokens']
inputs = tokenizer(inputs, is_split_into_words=True)
print(inputs.tokens())
```

```python
def align_labels_with_tokens(labels, word_ids):
  new_labels = []
  current_word=None
  for word_id in word_ids:
    if word_id != current_word:
      current_word = word_id
      label = -100 if word_id is None else labels[word_id]
      new_labels.append(label)

    elif word_id is None:
      new_labels.append(-100)

    else:
      label = labels[word_id]

      if label%2==1:
        label = label + 1
      new_labels.append(label)

  return new_labels
```

```python
labels = data['train'][0]['ner_tags']
word_ids = inputs.word_ids()
print(labels, word_ids)
```

```python
align_labels_with_tokens(labels, word_ids)
```

```python
def tokenize_and_align_labels(examples):
  tokenized_inputs = tokenizer(examples['tokens'], truncation=True, is_split_into_words=True)

  all_labels = examples['ner_tags']

  new_labels = []
  for i, labels in enumerate(all_labels):
    word_ids = tokenized_inputs.word_ids(i)
    new_labels.append(align_labels_with_tokens(labels, word_ids))

  tokenized_inputs['labels'] = new_labels

  return tokenized_inputs
```

```python
tokenized_datasets = data.map(tokenize_and_align_labels, batched=True, remove_columns=data['train'].column_names)
```

```python
tokenized_datasets
```

```python
from transformers import DataCollatorForTokenClassification

data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)
```

```python
batch = data_collator([tokenized_datasets['train'][i] for i in range(2)])
batch
```

```python
import evaluate
metric = evaluate.load('seqeval')
```

```python
ner_feature = data['train'].features['ner_tags']
ner_feature
```

```python
label_names = ner_feature.feature.names
label_names
```

```python
labels = data['train'][0]['ner_tags']
labels = [label_names[i] for i in labels]
labels
```

```python
predictions = labels.copy()
predictions[2] = "O"

metric.compute(predictions=[predictions], references=[labels])
```

```python
import numpy as np

def compute_metrics(eval_preds):
  logits, labels = eval_preds

  predictions = np.argmax(logits, axis=-1)

  true_labels = [[label_names[l] for l in label if l!=-100] for label in labels]

  true_predictions = [[label_names[p] for p,l in zip(prediction, label) if l!=-100]
                      for prediction, label in zip(predictions, labels)]

  all_metrics = metric.compute(predictions=true_predictions, references=true_labels)

  return {"precision": all_metrics['overall_precision'],
          "recall": all_metrics['overall_recall'],
          "f1": all_metrics['overall_f1'],
          "accuracy": all_metrics['overall_accuracy']}
```

```python
id2label = {i:label for i, label in enumerate(label_names)}
label2id = {label:i for i, label in enumerate(label_names)}
```

```python
from transformers import AutoModelForTokenClassification

model = AutoModelForTokenClassification.from_pretrained(
                                                    model_checkpoint,
                                                    id2label=id2label,
                                                    label2id=label2id)
```

```python
from transformers import TrainingArguments

args = TrainingArguments("distilbert-finetuned-ner",
                         evaluation_strategy = "epoch",
                         save_strategy="no",
                         learning_rate = 1e-6,
                         num_train_epochs=30,
                         weight_decay=0.01,
                         logging_steps=100,
                         report_to="none",
                         )
```

```python
from transformers import Trainer
trainer = Trainer(model=model,
                  args=args,
                  train_dataset = tokenized_datasets['train'],
                  eval_dataset = tokenized_datasets['test'],
                  data_collator=data_collator,
                  compute_metrics=compute_metrics,
                  tokenizer=tokenizer)

trainer.train()
trainer.save_model('distilbert-finetuned-ner')
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, auc
import sns
import matplotlib.pyplot as plt
preds = trainer.predict(tokenized_datasets['eval'])
cm = confusion_matrix(preds.label_ids.flatten(), preds.predictions.argmax(axis=-1).flatten())
sns.heatmap(cm, annot=True, fmt='d')
plt.title('Confusion matrix')
plt.show()

fpr, tpr, _ = roc_curve(preds.label_ids.flatten(), preds.predictions[:,1])
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from matplotlib import pyplot as plt

def generate_confusion_matrix(trainer, dataset, label_names):
    # Get predictions and true labels
    predictions, labels, _ = trainer.predict(dataset)
    predictions = np.argmax(predictions, axis=-1)

    # Flatten predictions and labels, ignoring padding (-100)
    true_labels = [label for label_seq in labels for label in label_seq if label != -100]
    pred_labels = [pred for pred_seq, label_seq in zip(predictions, labels)
                   for pred, label in zip(pred_seq, label_seq) if label != -100]

    # Generate confusion matrix
    cm = confusion_matrix(true_labels, pred_labels, labels=list(range(len(label_names))))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_names)
    disp.plot(cmap="viridis", xticks_rotation="vertical")
    plt.title("Confusion Matrix")
    plt.show()

# Call the function
generate_confusion_matrix(trainer, tokenized_datasets['eval'], label_names)

from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

def plot_roc_curve(trainer, dataset, label_names):
    # Get predictions and true labels
    predictions, labels, _ = trainer.predict(dataset)
    predictions = np.argmax(predictions, axis=-1)

    # Flatten predictions and labels, ignoring padding (-100)
    true_labels = [label for label_seq in labels for label in label_seq if label != -100]
    pred_probs = [prob for prob_seq, label_seq in zip(predictions, labels)
                  for prob, label in zip(prob_seq, label_seq) if label != -100]

    # Binarize the labels for ROC computation
    num_classes = len(label_names)
    true_labels_bin = label_binarize(true_labels, classes=list(range(num_classes)))
    pred_probs_bin = label_binarize(pred_probs, classes=list(range(num_classes)))

    # Compute ROC curve and AUC for each class
    plt.figure(figsize=(10, 8))
    for i, label in enumerate(label_names):
        fpr, tpr, _ = roc_curve(true_labels_bin[:, i], pred_probs_bin[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{label} (AUC = {roc_auc:.2f})")

    plt.plot([0, 1], [0, 1], "k--", label="Random Guess")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve for Each Class")
    plt.legend(loc="lower right")
    plt.grid()
    plt.show()

# Call the function
plot_roc_curve(trainer, tokenized_datasets['eval'], label_names)
```
